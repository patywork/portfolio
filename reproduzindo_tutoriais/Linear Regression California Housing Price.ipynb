## Implementação da Regressão Linear ##

Este notebook apenas trás como implementar a Regressão Linear no Stats e no Sckit_Learn 
Nao tem nenhum tratamento nos dados que utilizamos para a implementação.

```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm
```

```python
data = pd.read_csv('housing.csv')
```

### StatsModel ###

```python
y= 'median_house_value'
```

```python
data_dropped=data.drop([y],axis=1) #retira a variável target(y)
```

```python
#porcentagem de NANs
data.isna().sum()/data.shape[0]
```

Como a idéia aqui é rodar o modelo, vamos dropar o que tiver total_bedrooms zerado para conseguir rodar a regressão

```python
#transforma dados categóricos em dummies para o modelo e dropa os valores nulos
data_dropped.dropna(inplace=True)
data_dummies=pd.get_dummies(data_dropped)
```

```python
modelo = sm.OLS(np.array(data[y]), data_dummies)
res=modelo.fit()
res.summary()
```

```python
#normalizando
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=data_dummies
normalized_x = scaler.fit_transform(x)
normalized_x= pd.DataFrame(normalized_x, columns=x.columns)

modelo = sm.OLS(np.array(data[y]), normalized_x)
res=modelo.fit()
res.summary()
```
